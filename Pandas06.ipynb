```python
import pandas as michael #variabel
import numpy as jackson
```

```python
n_row = 5   #membuat baris dan kolom 5 x 5
n_col = 5
col = tuple ('FGHIJ') #menamai kolom
df = michael.DataFrame(jackson.random.randint(1, 10, size=(n_row, n_col)),# perintah membuat tabel 5 x 5 dengan nama kolom FGHIJ dengan isi anka random 1 - 10
                      columns = col) 
df
```

```python
df.rename(columns={'H':'kerja'}) #perintah untuk mengubah 1 nama kolom
```

```python
df.rename(columns={'F':'nama', 'G':'alamat', 'H':'kota'}) # sama kaya diatas cuman lebih banyak
```
